```
!pip install transformers torch --quiet

from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "gpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(model_name)

prompt = 'once there was a lion in the green jungle'

input_ids = tokenizer.encode(prompt, return_tensors = 'pt')

output = model.generate(
    input_ids,
    max_length = 100,
    num_return_sequences = 1,
    do_sample=True,
    top_k=50,
    top_p=0.95,
    pad_token_id=tokenizer.eos_token_id
)

generated_text = tokenizer.decode(output[0], skip_specail_tokens = True)

print(generated_text)
```

```
from transformers import pipeline

generator = pipeline("text2text-generation", model="google/flan-t5-large")

# Zero-shot Generation
prompt_zero = "Write a short and creative story about a brave girl named Riya who saved her village from a monster using her courage and intelligence."
output_zero = generator(prompt_zero, max_new_tokens=100)[0]['generated_text']
print("Zero-shot:\n", output_zero)

# Few-shot Generation
prompt_few = """Example: A clever boy used a map to find hidden treasure.
Now write a story about a brave girl who saved her village."""
output_few = generator(prompt_few, max_new_tokens=100)[0]['generated_text']
print("\nFew-shot:\n", output_few)

# Chain-of-thought
prompt_cot = """Let's write a story step by step:
Step 1: A brave girl saw danger near her village.
Step 2:"""
output_cot = generator(prompt_cot, max_new_tokens=100)[0]['generated_text']
print("\nChain-of-Thought:\n", output_cot)
```

```
from transformers import BartTokenizer, BartForConditionalGeneration
import torch

# Load model and tokenizer
model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)

# Use GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# Input paragraph
text = """
Artificial Intelligence is changing the world...
"""

# Tokenize and generate summary+
inputs = tokenizer([text], max_length=1024, return_tensors="pt", truncation=True).to(device)
summary_ids = model.generate(inputs["input_ids"], num_beams=4, max_length=60, min_length=15, length_penalty=2.0)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print("📌 Summary:", summary)
```

```
qa = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

context = "The Sun is a star at the center of our solar system. It gives light and heat to Earth."
question = "Which gives light and heat to Earth?"

answer = qa(question=question, context=context)
print("\nQ&A:\n", answer['answer'])
```

```
!pip install transformers sentencepiece --quiet

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load tokenizer and model
model_name = "facebook/nllb-200-distilled-600M"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Input text and language codes
src_lang = "tel_Telu"       # Telugu
tgt_lang = "eng_Latn"       # English
text = "నాకు కొత్త విషయాలు నేర్చుకోవడం చాలా ఇష్టం."

# Tokenize input
inputs = tokenizer(text, return_tensors="pt")

# Set the target language using forced_bos_token_id
forced_bos_token_id = tokenizer.convert_tokens_to_ids(tgt_lang)

# Generate translation
outputs = model.generate(**inputs, forced_bos_token_id=forced_bos_token_id)
translated = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]

print("🔤 Translation:", translated)
```
